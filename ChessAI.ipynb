```python
#import
from __future__ import annotations
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd


from enum import Enum
from typing import List
```

```python
#Couleur
class Couleur(Enum):
    Noir = 0
    Blanc = 1
```

```python
#Categorie
class Categorie(Enum):
    Tour = 1
    Fou = 2
    Cavalier = 3
    Roi = 4
    Reine = 5
    Pion = 6
```

```python
#Piece
class Piece:
    categorie:Categorie
    couleur:Couleur
    x:int
    y:int
    plateau:Plateau
    
    def actualiserCoupsPossible(self):
        self.coupsPossible.clear()
        
        flag:bool
        coupsPossible:List[Case] = []
        x:int = 0
        y:int = 0
        mx:int = 0
        my:int = 0
        
        #Pion blanc
        if self.categorie == Categorie.Pion and self.couleur == Couleur.Blanc:

            # Avancer de 1
            if 0 <= self.x < 8 and 0 <= self.y + 1 < 8:
                if self.plateau[self.x][self.y + 1].estOccupe == False:
                    self.coupsPossible.append(self.plateau[self.x][self.y + 1])

            # Avancer de 2
            if self.y == 1:
                if 0 <= self.x < 8 and 0 <= self.y + 2 < 8:
                    if self.plateau[self.x][self.y + 2].estOccupe == False:
                        self.coupsPossible.append(self.plateau[self.x][self.y + 2])

            # Capture gauche
            if 0 <= self.x - 1 < 8 and 0 <= self.y + 1 < 8:
                if self.plateau[self.x - 1][self.y + 1].estOccupe == True and \
                   self.plateau[self.x - 1][self.y + 1].occupePar.couleur != self.couleur:
                    self.coupsPossible.append(self.plateau[self.x - 1][self.y + 1])

            # Capture droite
            if 0 <= self.x + 1 < 8 and 0 <= self.y + 1 < 8:
                if self.plateau[self.x + 1][self.y + 1].estOccupe == True and \
                   self.plateau[self.x + 1][self.y + 1].occupePar.couleur != self.couleur:
                    self.coupsPossible.append(self.plateau[self.x + 1][self.y + 1])
                            
        #Pion noir            
        if self.categorie == Categorie.Pion and self.couleur == Couleur.Noir:

            if 0 <= self.x < 8 and 0 <= self.y - 1 < 8:
                if self.plateau[self.x][self.y - 1].estOccupe == False:
                    self.coupsPossible.append(self.plateau[self.x][self.y - 1])

            if self.y == 6:
                if 0 <= self.x < 8 and 0 <= self.y - 2 < 8:
                    if self.plateau[self.x][self.y - 2].estOccupe == False:
                        self.coupsPossible.append(self.plateau[self.x][self.y - 2])

            if 0 <= self.x - 1 < 8 and 0 <= self.y - 1 < 8:
                if self.plateau[self.x - 1][self.y - 1].estOccupe == True and \
                   self.plateau[self.x - 1][self.y - 1].occupePar.couleur != self.couleur:
                    self.coupsPossible.append(self.plateau[self.x - 1][self.y - 1])

            if 0 <= self.x + 1 < 8 and 0 <= self.y - 1 < 8:
                if self.plateau[self.x + 1][self.y - 1].estOccupe == True and \
                   self.plateau[self.x + 1][self.y - 1].occupePar.couleur != self.couleur:
                    self.coupsPossible.append(self.plateau[self.x + 1][self.y - 1])
        
        #Tour
        if self.categorie == Categorie.Tour:
            x = 1
            while x < 8:
                if 0 <= self.x + x < 8:
                    if self.plateau[self.x + x][self.y].estOccupe == False:
                        self.coupsPossible.append(self.plateau[self.x + x][self.y])
                    else:
                        if self.plateau[self.x + x][self.y].occupePar.couleur != self.couleur:
                            self.coupsPossible.append(self.plateau[self.x + x][self.y])
                        break
                else:
                    break
                x += 1

            mx = 1
            while mx < 8:
                if 0 <= self.x - mx < 8:
                    if self.plateau[self.x - mx][self.y].estOccupe == False:
                        self.coupsPossible.append(self.plateau[self.x - mx][self.y])
                    else:
                        if self.plateau[self.x - mx][self.y].occupePar.couleur != self.couleur:
                            self.coupsPossible.append(self.plateau[self.x - mx][self.y])
                        break
                else:
                    break
                mx += 1

            y = 1
            while y < 8:
                if 0 <= self.y + y < 8:
                    if self.plateau[self.x][self.y + y].estOccupe == False:
                        self.coupsPossible.append(self.plateau[self.x][self.y + y])
                    else:
                        if self.plateau[self.x][self.y + y].occupePar.couleur != self.couleur:
                            self.coupsPossible.append(self.plateau[self.x][self.y + y])
                        break
                else:
                    break
                y += 1

            my = 1
            while my < 8:
                if 0 <= self.y - my < 8:
                    if self.plateau[self.x][self.y - my].estOccupe == False:
                        self.coupsPossible.append(self.plateau[self.x][self.y - my])
                    else:
                        if self.plateau[self.x][self.y - my].occupePar.couleur != self.couleur:
                            self.coupsPossible.append(self.plateau[self.x][self.y - my])
                        break
                else:
                    break
                my += 1
        
        #Fou
        if self.categorie == Categorie.Fou:
            # ↗ diagonale haut droite
            x = 1
            y = 1
            while x < 8 and y < 8:
                if 0 <= self.x + x < 8 and 0 <= self.y + y < 8:
                    if self.plateau[self.x + x][self.y + y].estOccupe == False:
                        self.coupsPossible.append(self.plateau[self.x + x][self.y + y])
                    else:
                        if self.plateau[self.x + x][self.y + y].occupePar.couleur != self.couleur:
                            self.coupsPossible.append(self.plateau[self.x + x][self.y + y])
                        break
                else:
                    break
                x += 1
                y += 1


            # ↙ diagonale bas gauche
            mx = 1
            my = 1
            while mx < 8 and my < 8:
                if 0 <= self.x - mx < 8 and 0 <= self.y - my < 8:
                    if self.plateau[self.x - mx][self.y - my].estOccupe == False:
                        self.coupsPossible.append(self.plateau[self.x - mx][self.y - my])
                    else:
                        if self.plateau[self.x - mx][self.y - my].occupePar.couleur != self.couleur:
                            self.coupsPossible.append(self.plateau[self.x - mx][self.y - my])
                        break
                else:
                    break
                mx += 1
                my += 1


            # ↖ diagonale haut gauche
            mx = 1
            y = 1
            while mx < 8 and y < 8:
                if 0 <= self.x - mx < 8 and 0 <= self.y + y < 8:
                    if self.plateau[self.x - mx][self.y + y].estOccupe == False:
                        self.coupsPossible.append(self.plateau[self.x - mx][self.y + y])
                    else:
                        if self.plateau[self.x - mx][self.y + y].occupePar.couleur != self.couleur:
                            self.coupsPossible.append(self.plateau[self.x - mx][self.y + y])
                        break
                else:
                    break
                mx += 1
                y += 1


            # ↘ diagonale bas droite
            x = 1
            my = 1
            while x < 8 and my < 8:
                if 0 <= self.x + x < 8 and 0 <= self.y - my < 8:
                    if self.plateau[self.x + x][self.y - my].estOccupe == False:
                        self.coupsPossible.append(self.plateau[self.x + x][self.y - my])
                    else:
                        if self.plateau[self.x + x][self.y - my].occupePar.couleur != self.couleur:
                            self.coupsPossible.append(self.plateau[self.x + x][self.y - my])
                        break
                else:
                    break
                x += 1
                my += 1
        
       # Cavalier
        if self.categorie == Categorie.Cavalier:
            try:
                if 0 <= self.x - 1 < 8 and 0 <= self.y + 2 < 8:
                    if self.plateau[self.x - 1][self.y + 2].estOccupe == False or \
                       self.plateau[self.x - 1][self.y + 2].occupePar.couleur != self.couleur:
                        self.coupsPossible.append(self.plateau[self.x - 1][self.y + 2])
            except IndexError:
                pass

            try:
                if 0 <= self.x + 1 < 8 and 0 <= self.y + 2 < 8:
                    if self.plateau[self.x + 1][self.y + 2].estOccupe == False or \
                       self.plateau[self.x + 1][self.y + 2].occupePar.couleur != self.couleur:
                        self.coupsPossible.append(self.plateau[self.x + 1][self.y + 2])
            except IndexError:
                pass

            try:
                if 0 <= self.x + 2 < 8 and 0 <= self.y + 1 < 8:
                    if self.plateau[self.x + 2][self.y + 1].estOccupe == False or \
                       self.plateau[self.x + 2][self.y + 1].occupePar.couleur != self.couleur:
                        self.coupsPossible.append(self.plateau[self.x + 2][self.y + 1])
            except IndexError:
                pass

            try:
                if 0 <= self.x + 2 < 8 and 0 <= self.y - 1 < 8:
                    if self.plateau[self.x + 2][self.y - 1].estOccupe == False or \
                       self.plateau[self.x + 2][self.y - 1].occupePar.couleur != self.couleur:
                        self.coupsPossible.append(self.plateau[self.x + 2][self.y - 1])
            except IndexError:
                pass

            try:
                if 0 <= self.x + 1 < 8 and 0 <= self.y - 2 < 8:
                    if self.plateau[self.x + 1][self.y - 2].estOccupe == False or \
                       self.plateau[self.x + 1][self.y - 2].occupePar.couleur != self.couleur:
                        self.coupsPossible.append(self.plateau[self.x + 1][self.y - 2])
            except IndexError:
                pass

            try:
                if 0 <= self.x - 1 < 8 and 0 <= self.y - 2 < 8:
                    if self.plateau[self.x - 1][self.y - 2].estOccupe == False or \
                       self.plateau[self.x - 1][self.y - 2].occupePar.couleur != self.couleur:
                        self.coupsPossible.append(self.plateau[self.x - 1][self.y - 2])
            except IndexError:
                pass

            try:
                if 0 <= self.x - 2 < 8 and 0 <= self.y - 1 < 8:
                    if self.plateau[self.x - 2][self.y - 1].estOccupe == False or \
                       self.plateau[self.x - 2][self.y - 1].occupePar.couleur != self.couleur:
                        self.coupsPossible.append(self.plateau[self.x - 2][self.y - 1])
            except IndexError:
                pass

            try:
                if 0 <= self.x - 2 < 8 and 0 <= self.y + 1 < 8:
                    if self.plateau[self.x - 2][self.y + 1].estOccupe == False or \
                       self.plateau[self.x - 2][self.y + 1].occupePar.couleur != self.couleur:
                        self.coupsPossible.append(self.plateau[self.x - 2][self.y + 1])
            except IndexError:
                pass

        if self.categorie == Categorie.Roi:
            pass #coups possible pour roi blanc
        if self.categorie == Categorie.Reine:
            pass #coups possible pour reine blanche
            
    def icon(self):
        return self.categorie.name[0] + self.categorie.name[1]        
    
    def __init__(self, categorie:Categorie, couleur:Couleur, x:int, y:int, plateau:Plateau):
        self.categorie = categorie
        self.couleur = couleur
        self.x = x
        self.y = y
        self.plateau = plateau
        self.coupsPossible:List[Case] = []
```

```python
#Joueur
class Joueur:
    couleur:Couleur
    plateau:Plateau
    
    def __init__(self, couleur:Couleur):
        self.couleur = couleur
        self.pieces:List[Piece] = []
        self.coupsPossible = []
    
    def lierPlateau(self, plateau:Plateau):
        self.plateau = plateau
        
    def definirPieces(self):
        self.pieces.clear()
        for y in range (8):
            for x in range (8):
                if self.plateau[x][y].estOccupe:
                    if self.plateau[x][y].occupePar.couleur == self.couleur:
                        self.pieces.append(self.plateau[x][y].occupePar)
                        
    def actualiserCoupsPossible(self):
        self.coupsPossible.clear()
        self.definirPieces()
        for piece in self.pieces:
            piece.actualiserCoupsPossible()
            self.coupsPossible.append((piece, piece.coupsPossible))
            
    def afficherCoupsPossible(self):
        self.actualiserCoupsPossible()
        for infoPiece in self.coupsPossible:
            print(infoPiece[0].categorie.name + " (" + str(infoPiece[0].x) + " " + str(infoPiece[0].y) + "):")
            for coup in infoPiece[1]:
                print("   -" + str(coup.x) + " " + str(coup.y))
```

```python
#Case
class Case:
    estOccupe:bool = False
    occupePar:Piece = None
    x:int
    y:int
    
    def __init__(self, estOccupe:bool, occupePar:Piece, x:int, y:int):
        self.estOccupe = estOccupe
        self.occupePar = occupePar
        self.x = x
        self.y = y
```

```python
#Plateau
class Plateau:
    joueurNoir:Joueur
    JoueurBlanc:Joueur
    
    def __init__(self, joueurNoir:Joueur, joueurBlanc:Joueur):
        self.plateau:List[List[Case]] = [
            [Case(False, None, x, y) for y in range(8)] for x in range(8)
        ]
        
        self.joueurNoir = joueurNoir
        self.joueurBlanc = joueurBlanc
        self.joueurNoir.lierPlateau(self.plateau)
        self.joueurBlanc.lierPlateau(self.plateau)
        
        self.remplirPlateau()
        
        self.joueurNoir.definirPieces()
        self.joueurBlanc.definirPieces()
        
    def remplirPlateau(self):
        for y in range (8):
            if y == 7:
                self.plateau[0][y] = Case(True, Piece(Categorie.Tour, Couleur.Noir, 0, y, self.plateau), 0, y)
                self.plateau[1][y] = Case(True, Piece(Categorie.Cavalier, Couleur.Noir, 1, y, self.plateau), 1, y)
                self.plateau[2][y] = Case(True, Piece(Categorie.Fou, Couleur.Noir, 2, y, self.plateau), 2, y)
                self.plateau[3][y] = Case(True, Piece(Categorie.Reine, Couleur.Noir, 3, y, self.plateau), 3, y)
                self.plateau[4][y] = Case(True, Piece(Categorie.Roi, Couleur.Noir, 4, y, self.plateau), 4, y)
                self.plateau[5][y] = Case(True, Piece(Categorie.Fou, Couleur.Noir, 5, y, self.plateau), 5, y)
                self.plateau[6][y] = Case(True, Piece(Categorie.Cavalier, Couleur.Noir, 6, y, self.plateau), 6, y)
                self.plateau[7][y] = Case(True, Piece(Categorie.Tour, Couleur.Noir, 7, y, self.plateau), 7, y)
            if y == 6:
                for x in range (8):
                    self.plateau[x][y] = Case(True, Piece(Categorie.Pion, Couleur.Noir, x, y, self.plateau), x, y)
            if y == 1:
                for x in range (8):
                    self.plateau[x][y] = Case(True, Piece(Categorie.Pion, Couleur.Blanc, x, y, self.plateau), x, y)
            if y == 0:
                self.plateau[0][y] = Case(True, Piece(Categorie.Tour, Couleur.Blanc, 0, y, self.plateau), 0, y)
                self.plateau[1][y] = Case(True, Piece(Categorie.Cavalier, Couleur.Blanc, 1, y, self.plateau), 1, y)
                self.plateau[2][y] = Case(True, Piece(Categorie.Fou, Couleur.Blanc, 2, y, self.plateau), 2, y)
                self.plateau[3][y] = Case(True, Piece(Categorie.Reine, Couleur.Blanc, 3, y, self.plateau), 3, y)
                self.plateau[4][y] = Case(True, Piece(Categorie.Roi, Couleur.Blanc, 4, y, self.plateau), 4, y)
                self.plateau[5][y] = Case(True, Piece(Categorie.Fou, Couleur.Blanc, 5, y, self.plateau), 5, y)
                self.plateau[6][y] = Case(True, Piece(Categorie.Cavalier, Couleur.Blanc, 6, y, self.plateau), 6, y)
                self.plateau[7][y] = Case(True, Piece(Categorie.Tour, Couleur.Blanc, 7, y, self.plateau), 7, y)
    
    def afficher(self):
        for x in range (7,-1,-1):
            for y in range(8):
                if (self.plateau[y][x].occupePar != None):
                    texte = self.plateau[y][x].occupePar.icon() + "  "
                else:
                    texte = "__  "
                print(texte, end ="")
            print("\n")
```
