```python
#import
from __future__ import annotations
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd


from enum import Enum
from typing import List
```

```python
#Couleur
class Couleur(Enum):
    Noir = 0
    Blanc = 1
```

```python
#Categorie
class Categorie(Enum):
    Tour = 1
    Fou = 2
    Cavalier = 3
    Roi = 4
    Reine = 5
    Pion = 6
```

```python
#Piece
class Piece:
    categorie:Categorie
    couleur:Couleur
    x:int
    y:int
    coupsPossible:List[Case] = []
    
    def rafraichirCoupsPossible(self, plateau:Plateau):
        coupsPossible:List[Case] = []
        x:int = 0
        y:int = 0
        mx:int = 0
        my:int = 0
        
        if self.couleur == Couleur.Blanc:
            if self.categorie == Categorie.Tour:
                pass #coups possible pour tour blanche
            if self.categorie == Categorie.Fou:
                pass #coups possible pour fou blanc
            if self.categorie == Categorie.Cavalier:
                pass #coups possible pour cavalier blanc
            if self.categorie == Categorie.Roi:
                pass #coups possible pour roi blanc
            if self.categorie == Categorie.Reine:
                pass #coups possible pour reine blanche
            if self.categorie == Categorie.Pion:
                try: #première case devant
                    if plateau[self.x][self.y + 1].estOccupe == false:
                        coupsPossile.append(plateau[self.x][self.y + 1])
                        try: #deuxième case devant si la première case devant est libre
                            if plateau[self.x][self.y + 2].estOccupe == false:
                                coupsPossile.append(plateau[self.x][self.y + 2])
                        except IndexError:
                            pass
                except IndexError:
                    pass
                
                try: #case avant geuche
                    if plateau[self.x - 1][self.y + 1].estOccupe == false:
                        if plateau[self.x - 1][self.y + 1].occupePar.couleur == Couleur.Noir:
                            coupsPossile.append(plateau[self.x - 1][self.y + 1])
                except IndexError:
                    pass
                try: #case avant droite
                    if plateau[self.x + 1][self.y + 1].estOccupe == false:
                        if plateau[self.x + 1][self.y + 1].occupePar.couleur == Couleur.Noir:
                            coupsPossile.append(plateau[self.x + 1][self.y + 1])
                except IndexError:
                    pass
                            
                    
        if self.couleur == Couleur.Noir:
            
            if self.categorie == Categorie.Tour:
                pass #coups possible pour tour noire
            if self.categorie == Categorie.Fou:
                pass #coups possible pour fou noir
            if self.categorie == Categorie.Cavalier:
                pass #coups possible pour cavalier noir
            if self.categorie == Categorie.Roi:
                pass #coups possible pour roi noir
            if self.categorie == Categorie.Reine:
                pass #coups possible pour reine noire
            if self.categorie == Categorie.Pion:
                try: #première case devant
                    if plateau[self.x][self.y - 1].estOccupe == false:
                        coupsPossile.append(plateau[self.x][self.y - 1])
                        try: #deuxième case devant si la première case devant est libre
                            if plateau[self.x][self.y - 2].estOccupe == false:
                                coupsPossile.append(plateau[self.x][self.y - 2])
                        except IndexError:
                            pass
                except IndexError:
                    pass
                
                try: #case avant geuche
                    if plateau[self.x - 1][self.y - 1].estOccupe == false:
                        if plateau[self.x - 1][self.y - 1].occupePar.couleur == Couleur.Noir:
                            coupsPossile.append(plateau[self.x - 1][self.y - 1])
                except IndexError:
                    pass
                try: #case avant droite
                    if plateau[self.x + 1][self.y - 1].estOccupe == false:
                        if plateau[self.x + 1][self.y - 1].occupePar.couleur == Couleur.Noir:
                            coupsPossile.append(plateau[self.x + 1][self.y - 1])
                except IndexError:
                    pass
            
            
    
    def __init__(self, categorie:Categorie, couleur:Couleur, x:int, y:int):
        self.categorie = categorie
        self.couleur = couleur
        self.x = x
        self.y = y
        
    def icon(self):
        return self.categorie.name[0] + self.categorie.name[1]
```

```python
#Joueur
class Joueur:
    couleur:Couleur
    pieces:List[Piece] = []
    
    def __init__(self, couleur:Couleur):
        self.couleur = couleur
        
    def DefinirPieces(self, plateau:Plateau):
        for y in range (8):
            for x in range (8):
                if plateau[x][y].estOccupe:
                    if plateau[x][y].occupePar.couleur == self.couleur:
                        self.pieces.append(plateau[x][y].occupePar)
```

```python
#Case
class Case:
    estOccupe:bool = False
    occupePar:Piece = None
    
    def __init__(self, estOccupe:bool, occupePar:Piece):
        self.estOccupe = estOccupe
        self.occupePar = occupePar
```

```python
#Plateau
class Plateau:
    
    plateau: List[List[Case]] = [
        [Case(False,None) for x in range(8)] for y in range(8)
    ]
    joueurNoir:Joueur
    JoueurBlanc:Joueur
    
    def __init__(self, joueurNoir:Joueur, joueurBlanc:Joueur):
        self.joueurNoir = joueurNoir
        self.joueurBlanc = joueurBlanc
        
        self.remplirPlateau()
        
        self.joueurNoir.DefinirPieces(self.plateau)
        self.joueurBlanc.DefinirPieces(self.plateau)
        
    def remplirPlateau(self):
        for y in range (8):
            if y == 7:
                self.plateau[0][y] = Case(True, Piece(Categorie.Tour, Couleur.Blanc, 0, y))
                self.plateau[1][y] = Case(True, Piece(Categorie.Cavalier, Couleur.Blanc, 1, y))
                self.plateau[2][y] = Case(True, Piece(Categorie.Fou, Couleur.Blanc, 2, y))
                self.plateau[3][y] = Case(True, Piece(Categorie.Reine, Couleur.Blanc, 3, y))
                self.plateau[4][y] = Case(True, Piece(Categorie.Roi, Couleur.Blanc, 4, y))
                self.plateau[5][y] = Case(True, Piece(Categorie.Fou, Couleur.Blanc, 5, y))
                self.plateau[6][y] = Case(True, Piece(Categorie.Cavalier, Couleur.Blanc, 6, y))
                self.plateau[7][y] = Case(True, Piece(Categorie.Tour, Couleur.Blanc, 7, y))
            if y == 6:
                for x in range (8):
                    self.plateau[x][y] = Case(True, Piece(Categorie.Pion, Couleur.Blanc, x, y))
            if y == 1:
                for x in range (8):
                    self.plateau[x][y] = Case(True, Piece(Categorie.Pion, Couleur.Noir, x, y))
            if y == 0:
                self.plateau[0][y] = Case(True, Piece(Categorie.Tour, Couleur.Noir, 0, y))
                self.plateau[1][y] = Case(True, Piece(Categorie.Cavalier, Couleur.Noir, 1, y))
                self.plateau[2][y] = Case(True, Piece(Categorie.Fou, Couleur.Noir, 2, y))
                self.plateau[3][y] = Case(True, Piece(Categorie.Reine, Couleur.Noir, 3, y))
                self.plateau[4][y] = Case(True, Piece(Categorie.Roi, Couleur.Noir, 4, y))
                self.plateau[5][y] = Case(True, Piece(Categorie.Fou, Couleur.Noir, 5, y))
                self.plateau[6][y] = Case(True, Piece(Categorie.Cavalier, Couleur.Noir, 6, y))
                self.plateau[7][y] = Case(True, Piece(Categorie.Tour, Couleur.Noir, 7, y))
    
    def afficher(self):
        for y in range (8):
            for x in range(8):
                if (self.plateau[x][y].occupePar != None):
                    texte = self.plateau[x][y].occupePar.icon() + "  "
                else:
                    texte = "__  "
                print(texte, end ="")
            print("\n")
```

```python
#main
jn = Joueur(Couleur.Noir)
jb = Joueur(Couleur.Blanc)
p = Plateau(jn, jb)
p.afficher()
```
